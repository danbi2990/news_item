## Index

```python
def filter_president(x):
    for p in president:
        if p.lower() in x.lower():
            return True
    else:
        return False


def filter_system(x, word):
    if word.lower() in x.lower():
        return True
    else:
        return False


idx_president = systems['Body'].apply(filter_president)
idx_fptp = systems['System'].str.contains('FPTP')
idx_PR = systems['System'].str.contains('PR')
idx_tworound = systems['System'].apply(lambda x: filter_system(x, 'two-round'))
idx_party = systems['System'].apply(lambda x: filter_system(x, 'party list')) | systems['System'].apply(lambda x: filter_system(x, 'party-list'))
idx_list = systems['System'].apply(lambda x: filter_system(x, 'list'))
idx_pr = idx_PR | idx_list
idx_list_not_party = idx_list & ~idx_party

idx_check = idx_list_not_party
print_len_and_list(systems.loc[idx_check])
```
