```python
def process_review(review):
    review['contents'].fillna('', inplace=True)
    return review
```

# Load 'review'

```python
review = mg.read_mongo('club', 'review')
review = process_review(review)
review.head()
```

```python
review['site_name'].value_counts()
club = review.groupby('site_name').agg({'contents':'sum', 'total_count':'max'}).sort_values('total_count', ascending=False)
club = club.reset_index()
club.head()

# review.sort_values('total_count', ascending=False)
```

# Load 'club'

```python
club2 = mg.read_mongo('club', 'club')

club2 = process_club(club2)
print(len(club2))
club2.head()
```

```python
result = club[['site_name', 'total_count']].merge(club2[['name', 'region', 'type']], left_on='site_name', right_on='name')
result['region'] = result['region'].str.replace(' 클럽', '')
result
```
