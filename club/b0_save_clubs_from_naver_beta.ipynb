```python
import json
import requests
import pandas as pd

from urllib.parse import quote_plus
from pprint import pprint

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', None)

%load_ext autoreload
%autoreload 2
```

```python
def get_poi_from_naver_beta(query, page=1, display_count=10):
    '''query example: Í∞ïÎÇ® ÌÅ¥ÎüΩ'''
    import json
    import requests
    query = quote_plus(query)
    url = f'https://beta.map.naver.com/api/search?caller=pcweb&query={query}&type=place&page={page}&displayCount={display_count}&isAdult=false&searchCoord=126.99355465;37.53333020201292&isPlaceRecommendationReplace=true&lang=ko'
    r = requests.get(url)
    poi = json.loads(r.text)
    return poi


def get_total_count(query):
    '''
    page100: result.address = None
    result.place.totalCount
    result.place.list
    '''
    poi = get_poi_from_naver_beta(query)
    return poi['result']['place']['totalCount']


def get_all_list(query):
    total_count = get_total_count(query)
    # example: 2141
    display_count = 100
    cycle = total_count // display_count + 2
    result_list = []
    for c in range(1, cycle):
        poi = get_poi_from_naver_beta(query, c, display_count)
        result_list = result_list + poi.get('result', []).get('place', []).get('list', [])
    for r in result_list:
        r['region'] = query
    return result_list


def save_clubs_to_mongo(query):
    import src.mongos as mg
    result_list = get_all_list(query)

    mg.to_mongo(result_list, 'club', 'poi_naver_beta', dtype='list')
```
