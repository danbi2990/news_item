```python
def get_review(id_, start=1, display=10):
    '''
    review.maxItemCount
    review.total
    review.items
    '''
    import json
    import requests
    from urllib.parse import quote_plus

    url = f"https://store.naver.com/sogum/api/fsasReviews?exclude=&businessId={id_}&category=restaurant&start={start}&display={display}"
    r = requests.get(url)
    review = json.loads(r.text)
    return review


def get_total_count(id_):
    review = get_review(id_)
    return review.get('total', 0)


def get_cycle_count(id_):
    review = get_review(id_)
    return review.get('maxItemCount', 0)


def get_all_review(id_, site_name):
    total_count = get_total_count(id_)
    cycle_count = get_cycle_count(id_)
    display_count = 50
    cycle = cycle_count // display_count + 1
    # total: 731, cycle:15 (0~14) (1, 51, 101, ... 701)
    result_list = []
    for c in range(cycle):
        start = 1 + c * display_count
        review = get_review(id_, start, display_count)
        result_list = result_list + review.get('items', [])
    for r in result_list:
        r['site_id'] = id_
        r['site_name'] = site_name
        r['total_count'] = total_count
    return result_list


# review = get_review('31663844',)
# pprint(review)
# cycle_count = get_cycle_count('31663844',)
# reviews = get_all_review('31663844',)
# print(len(reviews))
# pprint(reviews)
```

```python
for idx, row in club[['id', 'name']].iterrows():
    id_ = row['id']
    name = row['name']
    reviews = get_all_review(id_, name)
    if reviews:
        mg.to_mongo(reviews, 'club', 'review', dtype='list')
#     print(id_, name)
```
