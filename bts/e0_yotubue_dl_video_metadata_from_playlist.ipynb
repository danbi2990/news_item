```python
from pprint import pprint

import youtube_dl

import src.mongos as mg

%load_ext autoreload
%autoreload 2
```

# Const

```python
playlist_id = 'PL_Cqw69_m_yzbMVGvQA8QWrL_HdVXJQX7'
db = 'youtube_comment'
collection = 'bts'
```

# Get video_entries from playlist

```python
ydl = youtube_dl.YoutubeDL({
    'simulate': True,
    'extract_flat': 'in_playlist',
    'forcejson': True
})

with ydl:
    result = ydl.extract_info(playlist_id)

pprint(result)
```

```python
entries = result['entries']; entries[:3]
```

# Get info

```python
ydl2 = youtube_dl.YoutubeDL({
    'simulate': True,
    'outtmpl': '%(comment_count)s.$(ext)s',
})

id_ = entries[0]['id']
url = 'https://www.youtube.com/watch?v=' + id_
result = ydl2.download([url])
print(result)

# result = ydl2.extract_info(entries[0]['id'], download=False)
# pprint(result)
```

```python
keys = ['dislike_count', 'duration', 'like_count', 'tags', 'thumbnail', 'track', 'upload_date', 'view_count']

for e in entries:
    result = ydl2.extract_info(e['id'], download=False)
    
    for k in keys:
        e[k] = result[k]

pprint(entries)
```

# to_mongo

```python
mg.to_mongo(entries, db, collection, dtype='list')
```
