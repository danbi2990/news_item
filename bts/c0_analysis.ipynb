# 언어별 비율

```python
def group_by_lang(df):
    total = len(df)
    lang = df.groupby('langCode').agg({'id': 'count'}).sort_values('id', ascending=False)
    lang.reset_index(inplace=True)

    lang = lang.merge(lang_table[['639-1', 'ISO language name']], left_on='langCode', right_on='639-1', how='left')
    lang.drop(['639-1'], axis=1, inplace=True)

    lang.rename(columns={'id':'count', 'ISO language name': 'langName'}, inplace=True)
    lang['rate'] = lang['count'] / total
    return lang

# 제외: de, do, tl, id, 
```

# 중복 유저

```python
def group_author(df):
    total = len(df)
    author = df.groupby('author').agg({'id': 'count'}).sort_values('id', ascending=False)
    author.reset_index(inplace=True)
    author.rename(columns={'id': 'count'}, inplace=True)
    author['rate'] = author['count'] / total
    return author
```

# v!3w

```python
# comments.loc[comments['text'].str.contains(r'v[~i][~e]w', regex=True), cols].shape
cols = ['author', 'dateTime2', 'id', 'langCode', 'likes', 'text']

def get_words(df, regex):
    view = df.loc[df['text'].str.contains(regex, regex=True), cols]
    return view
```

```python
view = get_words(comments, r'v\w\ww')

print(len(view))

print(len(view[view['text'].str.contains('view')]))

not_view = view[~view['text'].str.contains('view')]

print(len(not_view))
```

```python
not_view.head(8)['text']
```

```python
from collections import Counter

import re
not_view_words = re.findall(r'v\w\ww', ' '.join(not_view['text'].to_list()))
```

```python
not_view_cnt = Counter(not_view_words)
tmp = pd.DataFrame(not_view_cnt.most_common(), columns=['word', 'count'])
# tmp.to_csv(dir_tableau + 'words_view.tsv', sep='\t', index=False)
tmp
```

```python
stream = get_words(comments, r'str\w\wm')
print(len(stream))
```

```python
idx_stream = stream['text'].str.contains('stream')
print(len(stream[idx_stream]))
print(len(stream[~idx_stream]))
```

```python
not_stream = stream[~idx_stream]
not_stream_words = re.findall(r'str\w\wm', ' '.join(not_stream['text'].to_list()))
not_stream_cnt = Counter(not_stream_words)
```

```python
not_stream_cnt.most_common()

tmp = pd.DataFrame(not_stream_cnt.most_common(), columns=['word', 'count'])
# tmp.to_csv(dir_tableau + 'words_stream.tsv', sep='\t', index=False)
tmp
```
