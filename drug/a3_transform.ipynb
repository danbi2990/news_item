```python
def get_ids_from_file_name(file_name):
    file_name_wo_ext = file_name.split('.')[0]
    ids = file_name_wo_ext.split('_')
    return ids


def filter_year_table(file_name, table_condition):
    ids = get_ids_from_file_name(file_name)
    table = ids[2]
    return (table==table_condition)


def get_nth_files_from_files(files, n):
    return [f for f in files if filter_year_table(f, n)]
```

```python
def get_year_date_from_file_name(file_name):
    ids = get_ids_from_file_name(file_name)
    year, date = ids[0], ids[1]
    return f'{year}{date:>02}'


def get_df_from_file(file_name):
    file_path = f'{dir_proj}{file_name}'
    year_date = get_year_date_from_file_name(file_name)
    df = pd.read_csv(file_path, sep='\t')
    df['기간'] = year_date
    return df


def get_df_appended_from_files(files):
    df = pd.DataFrame()
    for file in files:
        tmp = get_df_from_file(file)
        df = df.append(tmp)
    return df


def get_df_from_df_appended(df, cols):
    df.rename(columns=cols, inplace=True)
    
    return df


def get_cols_12():
    col0 = ['대마', '마약', '향정', '합계']
    col1 = ['건수', '단속', '구속']
    cols = [f'{c0}_{c1}' for c0 in col0 for c1 in col1]
    return cols


def get_cols_8():
    col0 = ['대마', '마약', '향정', '합계']
    col1 = ['남', '여']
    cols = [f'{c0}_{c1}' for c0 in col0 for c1 in col1]
    return cols


def get_df_value_aligned(df):
    import re
    from io import StringIO
    
    csv = df.to_csv(index=False)
    csv = re.sub(r',{2,}', ',', csv)
    csv = csv.replace(' ', '","')
    
    csv_io = StringIO(csv)
    df = pd.read_csv(csv_io)
    return df
```

```python
def get_df_8_from_files(files):
    df = get_df_appended_from_files(files)
    
#     cols = get_cols_8()
#     columns = {f'{i+1}': c for i, c in enumerate(cols)}
    cols = ['대마', '마약', '향정', '합계']
    columns = {f'{i+2}': c for i, c in enumerate(cols)}
    
#     df.rename(columns=columns, inplace=True)
    df.rename(columns={**{'1': '지역'}, **columns}, inplace=True)
    
    idx_filter = df.index!=0
#     idx_empty = df['15미만'].isna()
    df = df.loc[idx_filter]
    
    df.loc[df.index==1, '지역'] = '합계'
#     df.loc[df['종류'].isna(), '종류'] = '합계'
    
    df.drop(columns=['0'], axis=1, inplace=True)
#     df.fillna('', inplace=True)
#     df = df.applymap(lambda x: x.split('\n')[0]).sort_values('기간', ascending=True)

    df['합계'] = df['합계'].apply(lambda x: x.split('\n')[0])
    df['지역'] = df['지역'].apply(lambda x: x.replace('\n', ''))
    df = df.sort_values('기간', ascending=True)
    
    return df


files_8 = get_nth_files_from_files(files, '8')
print(len(files_8))
df_8 = get_df_8_from_files(files_8)
df_8
```
