# Function: File

```python
def get_files(dir_):
    return [f for f in os.listdir(dir_) if '.csv' in f]


def validate_file(file):
    with open(file) as f:
        text = f.read()
    return not (('fail' in text) and ('error' in text))


def get_valid_files(dir_):
    return [f for f in get_files(dir_) if validate_file(dir_ + f)]


def get_hidden_region_from_file(file):
    return file.split('-')[1]


def get_downloaded_regions(dir_):
    return [get_hidden_region_from_file(file) for file in get_valid_files(dir_) if '.csv' in file]


dir_item = '/home/jake/Documents/News_Item/Browser/'
dir_arch = dir_item + 'archive/'
dir_arch2 = dir_item + 'archive2/'

files = get_valid_files(dir_item)
print(files[:2])
len(files)
```

# Validation & Load

```python
def get_df_all_from_files(files):
    df_all = {}
    for file in files:
        path_file = dir_item + file
#         print(path_file)
        is_valid = validate_file(path_file)
        if is_valid:
            hidden_region = get_hidden_region_from_file(file)
            try:
                df_all[hidden_region] = pd.read_csv(path_file)
            except:
                with open(path_file) as f:
                    errored_text = f.read()
                print(f"Error hidden region: {hidden_region}")
                print(errored_text)

    return df_all


df_all = get_df_all_from_files(files)
df_all['ww']
```

# Merge df_all

```python
def get_df_merged(df_all):
    df_merged = pd.DataFrame()
    for region_code, df in df_all.items():
        df_tmp = df
        df_tmp['Code'] = region_code
        df_tmp['Region'] = code2name[region_code]
        df_merged = df_merged.append(df_tmp)
    return df_merged
```

```python
df_merged = get_df_merged(df_all)
```
