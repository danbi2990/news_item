```python
def get_clean_file(row):
    dir_page = row['dir_page_fmt']
    file = row['file']
    return f'pdf/{dir_page}/cleaned_{file}'

failed = pd.read_csv('log.txt', sep='\t',)
failed['dir_page_fmt'] = failed['dir_page'].apply(lambda x: f'{x:>03}')
failed['clean_file'] = failed.apply(get_clean_file, axis=1)
failed.head()
```

```python
def run_mutool():
    import subprocess

    for idx, row in failed.iterrows():
        dir_page = row['dir_page_fmt']
        file = row['file']
        failed_file = f'pdf/{dir_page}/{file}'
        clean_file = row['clean_file']
        command = ['mutool', 'clean', f'{failed_file}', f'{clean_file}']
        subprocess.run(command)
```

```python
def save_tables(pdfs):
    '''
    ~/.virtualenvs/workspace/lib/python3.6/site-packages/PyPDF2/util.py
    238 line, latin-1 => utf8
    '''

    for idx, pdf in enumerate(pdfs):
        page, file_name = pdf.split('/')[1:3]
        dir_tables_page = f'tables/{page}/'
        try:
            table = camelot.read_pdf(pdf, pages='1-end')
            table.export(dir_tables_page + file_name.replace('.pdf', '.csv'), f='csv')
            print(idx, dir_tables_page + file_name.replace('.pdf', '.csv'), ' saved.')
        except:
            with open('log2.txt', 'a') as f:
                f.write(f'{pdf}\n')
            continue

# save_tables()
```
