## Pre-process

### Datetime Format

```python
def test_date_time(x):
    import datetime
    if isinstance(x, datetime.date):
        print('date')
    elif isinstance(x, datetime.datetime):
        print('datetime')
    else:
        print('nothing')

        
def get_date_time(x):
    # 3/21/2019 7:13:00
    import datetime
    format_ = '%m/%d/%Y %H:%M:%S'
    date_time = datetime.datetime.strptime(x, format_)
    return date_time


def get_date(x):
    # 1/29/2016
    import datetime
    format_ = '%m/%d/%Y'
    date = datetime.datetime.strptime(x, format_)
    return date
    

# geothermal['날짜'].apply(get_date_time)
# earthquake['발생시각'].apply(test_date_time)

# earthquake['발생시각'].apply(get_date_time)
# geothermal['날짜'].apply(get_date)

# earthquake['발생시각_dt'] = earthquake['발생시각'].apply(get_date_time)
# geothermal['날짜_dt'] = geothermal['날짜'].apply(get_date)
```

### Filling

```python
def transform_dash_to_zero(x):
    if x == '-':
        print('hi')
        return '0'
    else:
        return x
# geothermal
```

### Strip Columns

```python
def strip_columns(df):
    df.rename(columns={x: x.strip() for x in df.columns}, inplace=True)
```

### Process

```python
def process_raw_earthquake(raw_earthquake):
    earthquake = raw_earthquake.copy()
    earthquake['발생시각_dt'] = earthquake['발생시각'].apply(get_date_time)
    earthquake = earthquake.applymap(transform_dash_to_zero)
    strip_columns(earthquake)
    return earthquake


def process_raw_geothermal(raw_geothermal):
    geothermal = raw_geothermal.copy()
    geothermal['날짜_dt'] = geothermal['날짜'].apply(get_date)
    geothermal = geothermal.applymap(transform_dash_to_zero)
    strip_columns(geothermal)
    return geothermal
```

```python
earthquake = process_raw_earthquake(raw_earthquake)
geothermal = process_raw_geothermal(raw_geothermal)
```
